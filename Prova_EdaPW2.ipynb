```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtra il dataframe per le righe desiderate
filtered_df = df[(df['metric'] == 'Rate') & (df['measure'] == 'Deaths')].copy()

# Converti la colonna 'val' in tipo float
filtered_df['val'] = filtered_df['val'].astype(float)

# Converti la colonna 'year' in tipo numerico
filtered_df['year'] = pd.to_numeric(filtered_df['year'], errors='coerce')

# Rimuovi eventuali righe con 'NaN' nella colonna 'year'
filtered_df = filtered_df.dropna(subset=['year'])

# Crea intervalli di 5 anni utilizzando pd.cut
min_year = 1990
max_year = int(filtered_df['year'].max())
bins = range(min_year, max_year + 6, 5)
filtered_df.loc[:, 'year_group'] = pd.cut(filtered_df['year'], bins=bins, right=False)

# Estrai il valore medio dall'intervallo
filtered_df['year_group'] = filtered_df['year_group'].apply(lambda x: x.mid)

# Converti la colonna 'year_group' in tipo int
filtered_df['year_group'] = filtered_df['year_group'].astype(int)

# Crea l'istogramma con sns.barplot utilizzando 'continente' al posto di 'location'
plt.figure(figsize=(12, 8))
ax = sns.barplot(data=filtered_df, x='year_group', y='val', hue='continente', ci=None)

# Imposta i titoli e le etichette degli assi
plt.title('Distribuzione del valore "Rate" per Deaths ogni 5 anni')
plt.xlabel('Anno')
plt.ylabel('Valore "Rate"')

# Inserisci la legenda
plt.legend(title='Continente', bbox_to_anchor=(1.05, 1), loc='upper left')

# Aggiungi le cifre sopra le barre
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8)

# Visualizza il grafico
plt.show()


```
